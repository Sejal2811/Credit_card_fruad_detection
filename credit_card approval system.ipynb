```python
data = data[["Credit_Number","Car_Owner","Propert_Owner","Annual_income","EDUCATION","label"]]
```

```python
data
```

```python
from sklearn import preprocessing 
labelencoder = preprocessing.LabelEncoder()
```

```python
data["Car_Owner"]= labelencoder.fit_transform(data["Car_Owner"])
data["Propert_Owner"]= labelencoder.fit_transform(data["Propert_Owner"])
```

```python
data["EDUCATION"] = labelencoder.fit_transform(data["EDUCATION"])
```

```python
data
```

```python
# 10. Save the preprocessed data
data.to_csv('Approval.csv', index=False)
```

```python
X= data.iloc[: ,:-1]
```

```python
y= data.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
def modelperformance(predictions):
    print("Accuracy score on model {}". format(accuracy_score(y_test,predictions)))
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(scaled_X_train, y_train)
```

```python
import pickle

with open('approval_model.pkl', 'wb') as f:
    pickle.dump(log_model, f)
```

```python
log_predictions=log_model.predict(scaled_X_test)
```

```python
modelperformance(log_predictions)
```
